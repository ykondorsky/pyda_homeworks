## Задание 1
Дана переменная, в которой хранится слово из латинских букв. Напишите код, который выводит на экран:
- среднюю букву, если число букв в слове нечетное;
- две средних буквы, если число букв четное.

Примеры работы программы:  
1. `word = 'test'`

Результат:  
`es`
2.     
`word = 'testing'`

Результат:  
`t`

````python
my_word = 'test'
a = len(my_word)
if a % 2 == 0:
    print(my_word[(a//2)-1] + my_word[a//2])
else:
    print(my_word[a//2])
````

## Задание 2
Напишите программу, которая последовательно запрашивает у пользователя числа (по одному за раз) и после первого нуля выводит сумму всех ранее введенных чисел.

Примеры работы программы:  
1.   

```
Введите число:  
1

Введите число:  
4

Введите число:  
6

Введите число:  
0
```

Результат:  
`11`

2.      

```
Введите число:  
0
```  

Результат:  
`0`

````python
sum = 0
while True:
    a = int(input('Введите число: '))
    if a != 0:
        sum += a;
    else:
        break
print(sum)
````

## Задание 4

У нас есть список, содержащий информацию о среднедневной температуре в Фаренгейтах за произвольный период по странам (структура данных в примере).
Необходимо написать код, который рассчитает среднюю температуру за период в Цельсиях(!) для каждой страны.

Пример работы программы:  
```
countries_temperature = [
    ['Thailand', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
    ['Germany', [57.2, 55.4, 59, 59, 53.6]],
    ['Russia', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
    ['Poland', [50, 50, 53.6, 57.2, 55.4, 55.4]]
]
```

Результат:  
```
Средняя температура в странах:
Thailand  -  23.9 С
Germany  -  13.8 С
Russia  -  3.7 С
Poland  -  12.0 С
```

````python
countries_temperature = [
    ['Thailand', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
    ['Germany', [57.2, 55.4, 59, 59, 53.6]],
    ['Russia', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
    ['Poland', [50, 50, 53.6, 57.2, 55.4, 55.4]]
]
print('Средняя температура в странах:')
for i in range (0, len(countries_temperature)):
    temp_f = 0
    for j in range (0, len(countries_temperature[i][1])):
        temp_f += countries_temperature[i][1][j]
    temp_f_avg = temp_f / len(countries_temperature[i][1])
    temp_c_avg = (temp_f_avg - 32) * (5/9)
    print(countries_temperature[i][0] + ' - ' + "{:.1f}".format(temp_c_avg) + ' C')
````

## Задание 5 (необязательное)

Дан поток логов по количеству просмотренных страниц для каждого пользователя. Вам необходимо написать алгоритм, который считает среднее значение просмотров на пользователя. 
Т. е. надо посчитать отношение суммы всех просмотров к количеству уникальных пользователей.

Примеры работы программы:  
1.  

```
stream = [
    '2018-01-01,user1,3',
    '2018-01-07,user1,4',
    '2018-03-29,user1,1',
    '2018-04-04,user1,13',
    '2018-01-05,user2,7',
    '2018-06-14,user3,4',
    '2018-07-02,user3,10',
    '2018-03-21,user4,19',
    '2018-03-22,user4,4',
    '2018-04-22,user4,8',
    '2018-05-03,user4,9',
    '2018-05-11,user4,11',
]
```

Результат:  
`Среднее количество просмотров на уникального пользователя: 23.25`
2.     

```
stream = [
    '2018-01-01,user100,150',
    '2018-01-07,user99,205',
    '2018-03-29,user1001,81'
]
```

Результат:  
`Среднее количество просмотров на уникального пользователя: 145.33`

````python
stream = [
    '2018-01-01,user1,3',
    '2018-01-07,user1,4',
    '2018-03-29,user1,1',
    '2018-04-04,user1,13',
    '2018-01-05,user2,7',
    '2018-06-14,user3,4',
    '2018-07-02,user3,10',
    '2018-03-21,user4,19',
    '2018-03-22,user4,4',
    '2018-04-22,user4,8',
    '2018-05-03,user4,9',
    '2018-05-11,user4,11',
]
# составляем списки уникальных пользователей и количества просмотров для каждого из них (пока пустые)
users = []
views = []
for i in range(0, len(stream)):
    a = stream[i].split(',')
    # если юзера нет в списке - добавляем
    if not(a[1] in users):
        users.append(a[1])
        views.append(int(a[2]))
    # иначе добавляем количество просмотров для этого юзера в список
    else:
        views[users.index(a[1])] += int(a[2])
# считаем среднее количество просмотров
views_sum = 0
for i in range (0, len(views)):
    views_sum += views[i]
views_avg = views_sum / len(views)
print('Среднее количество просмотров на уникального пользователя: ' + "{:.2f}".format(views_avg))
    
````

## Задание 6 (необязательное)

Необходимо у пользователя запрашивать набор чисел разделенных пробелом. В результате должны выводиться числа, которые повторяются в вводе более одного раза.

Примеры работы программы:  
1.  

```
Введите числа:
4 8 0 3 4 2 0 3
```

Результат:  
`0 3 4`
2.     

```
Введите числа:
1 1 2 2 3 3
```

Результат:  
`1 2 3`
3.     

```
Введите числа:
1 1 1 1 1 2 2 2
```

Результат:  
`1 2`

````python
unique_numbers_list = []
user_input = input('Введите числа: ')
numbers_list = user_input.split(' ')
unique_numbers_list = []
output_list = []
for i in range (0, len(numbers_list)):
    # если число нам ещё не встречалось - добавляем в список уникальных чисел
    if not numbers_list[i] in unique_numbers_list:
        unique_numbers_list.append(numbers_list[i])
    # если число уже встречалось - оно должно быть в списке уникальных чисел, значит, добавляем его в список на вывод
    elif not numbers_list[i] in output_list:
        output_list.append(numbers_list[i])
output_list
    
````
