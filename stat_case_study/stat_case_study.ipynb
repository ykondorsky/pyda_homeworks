### Как критики относятся к спортивным играм?

Н0 - средний рейтинг критиков для спортивных игр >= 70 (положительно относятся) <br>
H1 - средний рейтинг < 70 <br>
Уровень значимости α примем за 5%

````python
from scipy import stats
alpha = 0.05
result = stats.ttest_1samp(game_data.loc[game_data.Genre=='Sports'].Critic_Score.dropna(), 70, alternative='less')

# средняя оценка спортивных игр
print(game_data.loc[game_data.Genre=='Sports'].Critic_Score.mean())
# результат t-теста
print(result)

if result.pvalue < alpha: 
    print('Отвергаем нулевую гипотезу, среднее меньше 70 (критики негативно относятся к спортивным играм)')
else:
    print('Не отвергаем нулевую нулевую гипотезу, критики положительно относятся к спортивным играм')
````

### Критикам нравятся больше игры на PC или на PS4?

Н0 - разницы нет (средняя оценка по выборкам одинакова) (критики одинаково относятся к играм на PC и PS4) <br>
H1 - есть статистически значимая разница между средними в выборках <br>
Уровень значимости α примем за 5%

````python
result = stats.ttest_ind(game_data.loc[game_data.Platform=='PC'].Critic_Score.dropna(),  
                         game_data.loc[game_data.Platform=='PS4'].Critic_Score.dropna())

# средние оценки игр на РС и PS4
print(game_data.loc[game_data.Platform=='PC'].Critic_Score.mean(), 
      game_data.loc[game_data.Platform=='PS4'].Critic_Score.mean())
# результаты t-теста
print(result)  

if result.pvalue < alpha: 
    print('Отвергаем нулевую гипотезу, критикам больше нравятся игры на {}'.format('PC' if result.statistic > 0 else 'PS4'))
else:
    print('Не отвергаем нулевую нулевую гипотезу, критики одинаково относятся к играм на РС и PS4')
````

### Критикам больше нравятся стрелялки или стратегии?

Н0 - разницы нет (средняя оценка по выборкам одинакова) (критики одинаково относятся к шутерам и стратегиям) <br>
H1 - есть статистически значимая разница между средними в выборках <br>
Уровень значимости α примем за 5%

````python
result = stats.ttest_ind(game_data.loc[game_data.Genre=='Shooter'].Critic_Score.dropna(),  
                         game_data.loc[game_data.Genre=='Strategy'].Critic_Score.dropna())

# средние оценки шутеров и стратегий
print(game_data.loc[game_data.Genre=='Shooter'].Critic_Score.mean(), 
      game_data.loc[game_data.Genre=='Strategy'].Critic_Score.mean())
# результаты t-теста
print(result)  

if result.pvalue < alpha: 
    print('Отвергаем нулевую гипотезу, критикам больше нравятся {}'.format('шутеры' if result.statistic > 0 else 'стратегии'))
else:
    print('Не отвергаем нулевую нулевую гипотезу, критики одинаково относятся к шутерам и стратегиям')
````

````python
spam_data = pd.read_csv('spam.csv')
spam_data
````

````python
import re

spam_data.Message = spam_data.Message.apply(lambda x: x.lower())   # приводим к нижнему регистру
spam_data.Message = spam_data.Message.apply(lambda x: re.sub('[\W_]+',' ', x))  # убираем знаки препинания
````

````python
# убираем стоп-слова
from nltk.corpus import stopwords
stopwords_set = set(stopwords.words('english'))
spam_data.Message = spam_data.Message.apply(lambda x: [word for word in x.split(' ') if word not in stopwords_set])
````

````python
# лемматизируем слова
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
spam_data.Message = spam_data.Message.apply(lambda x: ' '.join([wordnet_lemmatizer.lemmatize(word) for word in x]))
````

````python
# получаем матрицу TD-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(spam_data.Message.tolist())
names = tfidf.get_feature_names()
tfidf_matrix = pd.DataFrame(tfidf_matrix.toarray(), columns=names)
tfidf_matrix
````

````python
# разбиваем данные на тренировочную и тестовую выборку
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(tfidf_matrix, spam_data.Category, test_size=0.30, random_state=42)
````

````python
# создаём и обучаем модель
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score

lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
lda.predict(X_test)
accuracy_score(y_test, lda.predict(X_test))
````

Получаем хороший показатель точности - 0.967.

````python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, lda.predict(X_test))
````

Из матрицы ошибок видим, что 1618 (1445+173) из 1672 (размер тестовой выборки) значений были предсказаны корректно. 54 значения получены неверно. Представим их в таблице:

````python
mismatched = spam_data.loc[y_test.index, :].reset_index().join(pd.DataFrame({'Predicted': lda.predict(X_test)}))
mismatched = mismatched.loc[mismatched.Category != mismatched.Predicted]
mismatched[['Message', 'Category', 'Predicted']]
````

Из матрицы ошибок и из таблицы выше видим, что всего у нас 54 неверно классифицированных сообщений, из них 51 ложноотрицательных результатов (ошибок 2-ого рода): модель определила сообщение как не спам, хотя это был спам, и 3 ложноположительных.
